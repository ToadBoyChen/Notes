```python
import requests
from bs4 import BeautifulSoup

def print_character_grid(url):
    response = requests.get(url)
    if response.status_code != 200:
        print("Failed to retrieve data.")
        return
    
    soup = BeautifulSoup(response.text, 'html.parser')

    table = soup.find('table')

    rows = table.find_all('tr')
    
    character_data = []

    for row in rows[1:]:
        columns = row.find_all('td')
        if len(columns) >= 3:
            x_coord = columns[0].get_text(strip=True)
            character = columns[1].get_text(strip=True)
            y_coord = columns[2].get_text(strip=True)
            if x_coord and y_coord:
                character_data.append({
                    "x-coordinate": x_coord,
                    "character": character,
                    "y-coordinate": y_coord
                })
    
    max_x = -1
    max_y = -1

    for i in range(1, len(character_data)):
        current_entry = character_data[i]
        previous_entry = character_data[i - 1]
        current_x = int(current_entry["x-coordinate"])
        previous_x = int(previous_entry["x-coordinate"])
        current_y = int(current_entry["y-coordinate"])
        previous_y = int(previous_entry["y-coordinate"])

        if current_x > previous_x:
            max_x = max(max_x, current_x)
    
        if current_y > previous_y:
            max_y = max(max_y, current_y)
            
    MAX = max(max_x, max_y)

    grid = [[' ' for _ in range(MAX+ 1)] for _ in range(MAX + 1)]

    for entry in character_data:
        x = int(entry["x-coordinate"])
        y = int(entry["y-coordinate"])
        grid[max_y - y][x] = entry["character"]


    for row in grid:
        print(' '.join(row))


url = "https://docs.google.com/document/d/e/2PACX-1vQGUck9HIFCyezsrBSnmENk5ieJuYwpt7YHYEzeNJkIb9OSDdx-ov2nRNReKQyey-cwJOoEKUhLmN9z/pub"
print_character_grid(url)
```
