### Reporting

You must structure a report as such:

1. Title
1. Severity
    1. Informational - No real impact.
    1. Low - Business impact is minimal.
    1. Medium - Potential to cause harm to users, __does not reveal data__.
    1. High - Potential harm to reveal user data or aids exploitation of other vulnerabilities.
    1. Critical - High risk of personal data exposure, system compromise and/ or other severe impacts to business.
1. Description
1. Reproduction Steps
1. Impact
1. Mitigation
1. Affected Assets
