## Demon Games

Many though experiments in mathematics and physics involve demons, for some reason. This is the demon game. The rules are,

1. Read a statement left to right,
1. Whenever we see the forall symbol, $\forall$, the demon gets to choose. If there are restrictions, __the demon must obey__. 
1. Whenever there is a exists symbol, $\exists$, we get to choose. If there are restrictions, __we must obey__.
1. If the last statement is __true__, the __demon wins__, otherwise we win. 


Take the following specific game,

### Game One

The demon picks $n \in \N$, then we also pick an $m \in \N : m > n$ to win. You can clearly observe that we may choose to always win. 

### Game Two

the demon picks $x \in \R$ and picks $\epsilon \in \R : \epsilon > 0$. Then we pick $\delta \in \R : \delta > 0$. Then the demon once again picks $y \in \R : |x - y| < \delta$. We therefore win if $|x^2 - y^2| < \epsilon$. 

## Real Numbers

In order to understand the real numbers let us work into it. Recall that,

$$\begin{align*}
\N &= \{1,2,3,4,\dots\},\\
\N_0 &= \{0,1,2,3,\dots\},\\
\Z &= \{\dots,-1,0,1,\dots\},\\
\mathbb{Q} &= \left\{\frac{p}{q} : p,q\in\Z, q \neq 0\right\}.
\end{align*}$$

One should note that $\R$ contain all the sets prior, and all the prior sets contain their the prior sets of them. One should also note that all the sets are closed under addition and multiplication. only $\Z$ and $\mathbb{Q}$ is closed under subtraction and only $\mathbb{Q}$ is closed under division. Therefore $\R$ posses all the same qualities, and something unique.

Lastly, $\mathbb{Q}$ is a field, meaning that there is both an additive and multiplicative neutral, and a additive and multiplicative inverse.

#### Intervals

We define intervals as such,

$$\begin{align*}
[a, b] &:= \{x \in \R : a \leq x \leq b\},\\
(a, b) &:= \{ x\in \R : a < x , b\},\\
(a, b] &:= \{x\in\R:a<x\leq b\},\\
[a, b) &:= \{x\in\R:a\leq x < b\}.
\end{align*}$$

#### Maximum and Minimum 

Suppose $A \subseteq \R$. Then $x\in A$ is known as the maximum $\iff x\in A, ~ \forall y\in A: x \geq y$, and $x$ is the minimum if $x \in A, ~ \forall y \in A : x \leq y$. One should note that the maximum and minimum of a set is unique.

There is a difference between an upper/ lower bound and a maximum/ minimum. lets define it as such,

Let $A$ be a subset of $\R$. We say that for $x \in \R $  is the __upper bound__ of $A$ 

$$\iff \forall y \in A : x \geq y.$$

Similarly, for a lower bound, say $x \in \R$, we say $\iff \forall y \in A : x \leq y$.

We may notice that this means if we have a set, say $A = [0, 1]$, the maximum is $1$ but the upper bound could be $99999$ or something of such. Similarly, in the weird sense, if we have the set $A = (0, 1)$, then we have __no maximum__, as we may approach $1$ but we may never reach it. The upper bound therefore is $1$, as for all $y \in A : y \leq x$ for $x = 1$. Every number greater than $1$ is still an upper bound. 

We have seen that $A = \N$ has no maximum, but it __does have an upper bound__, which will be proven soon. 

#### Bounded Sets
A subset $A \in \R$ is said to be bounded above if it __has an upper bound__, or bounded below if it __has a lower bound__. We say that $A$ is bounded if its both bounded above and below.

#### Supremum and Infimum of Sets
Suppose $A \subseteq \R$. $x \in \R$ is the _least upper bound_ of $A$ iff $x$ is a upper bound of $A$ and every upper bound of $z$ for $A$ satisfies $x \leq z$. The least upper bound of $A$ is also known as the _supremum_, denoted by $\text{sup}(A)$. 

Conversely, $x \in \R$ is the _greatest lower bound_ of $A$ iff $x$ is a lower bound of $A$ and for every $z$ for $A$ satisfies $z \leq x$. This is known as the _infimum_ of $A$ and we denote it by $\text{inf}(A)$. 

In order to prove that a $x \in \R$ is the supremum, we must prove,

1. $\forall y \in A : y \leq x$,
1. $\forall z \in \R, z < x, \exists y \in A : z < y$.

Consider the following __Lemma__:

Consider a set $A \subseteq \R$. If $A$ has a maximum, then $\text{sup}(A) = \text{max}(A)$, else if $A$ has a supremum, then $A$ has a maximum $\iff \text{sup}(A) \in A$. If this holds, then $\text{max}(A) = \text{sup}(A)$.

### Definition (not formal) of Real Numbers
The real number set, $\R$ is defined by,
1. $\mathbb{Q} \subseteq \R$,
1. $\R$ is a field,
1. $\R$ is totally ordered, i.e., $\forall x, y \in \R$, exactly one is true: 
    1. $x > y$,
    1. $x = y$,
    1. $x < y$.
1. The _Completeness Axiom_, i.e., every non empty set of the real numbers which is bounded above has a __least upper bound__ in $\R$.

#### Remarks
1. $\mathbb{Q}$ satisfies (1), (2), (3),
1. $\mathbb{C}$ satisfies (1), (2),
1. The completeness axiom ensures that there are no gaps within $\R$. 
1. A field is a specialised group with two operations, $+$ and $\times$, where the following are satisfied,
    1. $a+(b + c) = (a + b) + c$, and $a \times (b \times c) = (a \times b) \times c$,
    1. $a + b = b + a$ and $a \times b = b \times a$,
    1. $\exists 1, 0 \in F : a + 0 = 0 + a = a$ and $a \times 1 = 1 \times a = a$,
    1. $\exists -a : a + (-a) = (-a) + a = 0$,
    1. $\exists a^{-1} : a \times a^{-1} = a^{-1} \times a = 1$,
    1. $a \times (b + c) = a \times b + a \times c$.

### Archimedian Property
For every real number $x$, there exists $n \in \N : n > x$.

_Proof_. Suppose $\exists x \in \R : \forall n \in \N : x \geq n$. If such $x$ exists, it is the upper bound for $\N$. Because $\N \subseteq \R$, the completeness axiom states that there must exist a supremum $y \in \R$. It then follows that $y - 1$ is not an upper bound for $\N$. $\exists n \in \N : n > y - 1$. But, $n + 1 > y$, but $y$ cannot be the upper bound for $\N$. 

#### Corollary
$\forall \epsilon \in \R, \epsilon > 0 ~ \exists n \in \N : \frac{1}{n} < \epsilon$.

_Proof_. We have $\frac{1}{\epsilon} \in \R$. By the archimedean property, $\exists n \in \N : \frac{1}{\epsilon} < n$. $\frac{1}{\epsilon} < n \iff 1 < n\epsilon \iff \frac{1}{n} < \epsilon$. 

#### Corollary
The set $\R$ is __dense__.

Suppose $x, y \in \R$ with $x < y$. Then $\exists z \in \mathbb{Q} : x < z < y$.

#### Theorem
The completeness axiom for infima. Suppose $A \subseteq \R$ is non empty and bounded below, i.e., $x \in \R : \forall y \in A, x < y$, then $A$ has a __greatest lower bound__.
