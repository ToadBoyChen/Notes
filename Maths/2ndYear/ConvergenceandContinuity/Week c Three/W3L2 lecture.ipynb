### Corollary - Irrationals are Dense in $\R$

Given $x, y \in \R, x > y, \exists z \in \R / \mathbb{Q} : x < z < y$.

_Proof_: Apply __theorem 217__: $\frac{x}{\sqrt{2}} < z < \frac{y}{\sqrt{2}}$. Its assumed that $z \in \mathbb{Q}$. We get $x < z\sqrt{2} < y$. We therefore know that $z\sqrt{2} \in \R/\mathbb{Q}$.

### Completeness Axiom for Infima

Suppose $a \subseteq \R, a \neq \empty : \exists ~ \text{infima} \in \R$.

_Proof Idea_: $B = \{-x : x \in A\}$. $A$ being non empty and bounded below means that $B$ is non empty and bounded above. Additionally, by the completeness axiom the $\text{sup}(B) = x \implies -x = \text{inf}(A)$. 

## Sequences

A sequence is just some list of numbers, for example $1,2,3,4,5,\dots$. __Thought: is there some rule to construct any given random list of numbers?__ For example can we find a rule to calculate $\pi, e, \sqrt{44}, 1, 99^3, \dots$?

You may think of a sequence as a function $f: \N \rightarrow \R$ by $n \mapsto x_n$.

If $(x_n)$ converges to zero, we may write $(x_n)\rightarrow 0$ or $\lim_{n \rightarrow \infty} (x_n) = 0$.

### Sequences Converging to Zero  

#### Definition

A sequence $(x_n)_{n = 1}^{\infty}$ converges to zero $\iff \forall \epsilon > 0 \exists N \in \N \forall n \in \N, n > N : |x_n| < \epsilon$.

To understand this definition lets play the __demon game__.

1. Demon picks $\epsilon > 0$.
1. We pick $N \in \N$. (This step is really important as there is an exists symbol!)
1. Demon picks $n > N$. 
1. Check if $|x_n| < \epsilon$.
    * We win if $|x_n| < \epsilon$.
    * We lose if $|x_n| \ge \epsilon$.

Play a Game involving $(x_n)_{n = 1}^{\infty} = (\frac{1}{n})_{n = 1}^{\infty}$.

1. Demon picks $\epsilon = \frac{1}{4}$
1. We pick $N = 10$.
1. Demon picks $n > 10$, say $n = 11$.
1. $|x_n| = \frac{1}{11} < \frac{1}{4} = \epsilon$. We win. 

Another game,

1. Demon picks $\epsilon = \frac{1}{1000}$
1. We pick $N = 1000$.
1. Demon picks $n > 1000$, say $n = 1001$.
1. $|x_n| = \frac{1}{1001} < \frac{1}{1000} = \epsilon$. We win.

The demon is basically saying that "I want $x_n$ to be within $\epsilon$ of $0$". "I want you to find how many $N$ along you must be to always be within $\epsilon$ of $0$."

Let us now suppose the demon picks some $\epsilon$, i.e., $\frac{1}{n} < epsilon$. $\exists N$ due to the Arch property. It makes sense to choose $N > \frac{1}{\epsilon}$ because $\epsilon \rightarrow \infty$, and so $N \rightarrow \infty$ as well, i.e., we need more steps to work within the smaller bounds chosen by $\epsilon$. If $n > N > \frac{1}{\epsilon} \implies \frac{1}{\epsilon}$. 

_Proof_: Claim - $\frac{1}{n}$ converges to zero. 

Given $\epsilon > 0$ take $N \in \N : N > \frac{1}{\epsilon}$. $\exists N$ because Archimedian property. Then given $n > N > \frac{1}{\epsilon} \implies n > \frac{1}{\epsilon}$ which is helpful because $\frac{1}{n} < \epsilon \implies |x_n| = |\frac{1}{n}| = \frac{1}{n} < \epsilon$.

#### Example 

$(x_n) = (\frac{1}{n^2})$. We want $|x_n| < \epsilon \iff \frac{1}{n^2} < \epsilon \iff n^2 > \frac{1}{\epsilon} \iff n > \frac{1}{\sqrt{\epsilon}}$.
