## Convergence to Zero

_Example_: $x_n = \frac{2\sin(n)}{n}$. Intuitively, this sequence oscillates about zero and tends to zero as the sine function is divided by a linearly increasing number. Now lets prove this using the definition of convergence to zero.

_Proof_: Given $\epsilon > 0,$ we [roughly] want $|x_n| = |\frac{2\sin n}{n}| = \frac{2}{n}|\sin n| \le \frac{2}{n}$. We now just need to show that $\frac{2}{n}$ is smaller than $\epsilon$. $\frac{2}{n} < \epsilon \iff n \ge \frac{2}{\epsilon}$. _Formally speaking_: $\epsilon > 0, N \in \N : N > \frac{2}{\epsilon}$, which exists by the Archimedean property of the natural numbers, we therefore say $n > N : |x_n| = \frac{2}{n}|\sin n| \le \frac{2}{n} < \frac{2}{N} < \epsilon$. Thus, $x_n \longrightarrow 0$ as $n \longrightarrow \infty$.

### Helpful Lemmas and Shortcuts for Proofs

#### Dominated Convergence

If $(x_n) \longrightarrow 0 $ and $(y_n) : |y_n| \le |x_n| \implies (y_n) \longrightarrow 0$. 

_Proof_: Given $\epsilon > 0$ and $(x_n) \longrightarrow 0$ then $\exists N \in \N : n > N: |x_n| < \epsilon$. For $n > N, |y_n| \le |x_n| < \epsilon \iff |y_n| < \epsilon$.

_Example_: $|\frac{1}{n^2 + n}| \le |\frac{1}{n}| \equiv |y_n| \le |x_n|$. By Dominated Convergence, $(y_n) \longrightarrow 0$.

#### Corollary of Dominated Convergence - Contrapositive

If $x_n \not{\longrightarrow} 0, |y_n| \ge |x_n| \forall n \in \N, (y_n) \not{\longrightarrow} 0$. 

#### Lemma - Substitute for Dominated Convergence

Suppose $(x_n) \longrightarrow 0$, and we fix $c \in \R$. Define $(y_n) = c(x_n) = cx_n \forall n \in \N$. Then $(y_n) \longrightarrow 0$.

_Proof for_ $c = 100$:  $\epsilon > 0, (x_n) \longrightarrow 0$. Take $\epsilon' = \frac{\epsilon}{100}$. Then $\exists N \in \N \forall n \in \N : |x_n| < \epsilon' = \frac{\epsilon}{100}$. $\implies |y_n| = 100|x_n| < 100(\frac{\epsilon}{100}) = \epsilon \iff |x_n| < \frac{\epsilon}{100}$.

#### Theorem - Summing Sequences

$(x_n) \longrightarrow 0, (y_n) \longrightarrow 0$. $(z_n) = x_n + y_n \forall n \in \N \implies (z_n) \longrightarrow 0$.

_Rough Proof_: $(x_n) \longrightarrow 0, (y_n) \longrightarrow 0 $. Prove $|z_n| < \epsilon$. 

The basic idea is: 

$$\begin{align*}
|z_n| &= |x_n + y_n| \le |x_n| + |y_n| [< \epsilon], \\
|x_n| + |y_n| &< \frac{1}{2}\epsilon + \frac{1}{2}\epsilon = \epsilon. \\
\end{align*}$$

_Real Proof_: $\forall \epsilon > o \exists N \in \N \forall n > N : |z_n| < \epsilon$. Define $\epsilon' = \frac{\epsilon}{2}$. Because $(x_n)\longrightarrow 0, \exists N \in \N : \forall n > N : |x_n| < \epsilon' = \frac{\epsilon}{2}$. Additionally because $(y_n)\longrightarrow 0, \exists M \in \N : n > M : |y_n| < \epsilon' = \frac{\epsilon}{2}$
. $N$ is how far we must go along for $x_n$ to be within $\epsilon'$, and additionally $M$ is how far we must go along for $y_n$ to be within $\epsilon'$.

Now, take $N' = \max\{N, M\}$. For $n > N'$, we have: $n > N \implies |x_n| < \epsilon'$, but also $n > M \implies |y_n| < \epsilon'$. Then $|z_n| = |x_n + y_n| \le |x_n| + |y_n| < \epsilon' + \epsilon' = \epsilon$. Thus, $(z_n) \longrightarrow 0$.

_Example_: Consider, $$x_n = \frac{3}{n^4} + \frac{4}{n + 1}. $$

We must show that these both converge to zero. $|\frac{3}{n^4}| \le 3|\frac{1}{n}|$, by dominated convergence, which shows that $|\frac{3}{n^4}| \longrightarrow 0$. Also $|\frac{4}{n + 1}| \le 4|\frac{1}{4}|$, which by dominated convergence converges to zero.

### Geometric Sequences

Take the example $(x_n) = \frac{1}{2^n} = \frac{1}{2}, \frac{1}{4}, \frac{1}{8}, \dots$. Our __claim__ is $(x_n)\longrightarrow 0$. The _proof_ is to show that $\frac{1}{2^n} \le \frac{1}{n} \forall n \in \N$, which then by dominated convergence will converge to zero. We will use __induction__. 

Base case: $n = 1$: $\frac{1}{2} \le 1$.

Inductive step: assume $\frac{1}{2^n} \le \frac{1}{n} \implies \frac{1}{2^{n+1}} \le \frac{1}{n+1}$.

Therefore $\frac{1}{2^{n+1}} = \frac{1}{2}\frac{1}{2^n} \le \frac{1}{2}\frac{1}{n} = \frac{1}{2n} = \frac{1}{n + n} \leq \frac{1}{n + 1}$. Proven. 

#### General Geometric Sequence 

$(x_n) = C^n, C\in \R$. If $|C| < 1$ then $(x_n) \longrightarrow 0$, else it does not. 

#### Lemma - Bernoullis Inequality

Let $\alpha \in \R : \alpha \ge -1$. Then $\forall n \in \N: (1+\alpha)^n \ge 1 + n\alpha$. _Proof_: Use induction.

_Proof_: By induction on $n$. Base case: $n = 1$: $(1+\alpha)^1 = 1 + \alpha \ge 1 + \alpha$. The inductive step: Assume that $(1 + \alpha)^n \ge 1 + n\alpha$ holds. $\implies (1 + \alpha)^{n + 1} = (1 + \alpha)(1 + \alpha)^n \ge (1 + \alpha)(1 + n\alpha) = 2 + n\alpha + \alpha + n\alpha^2 = 1 + (n+1)\alpha + n\alpha^2 \ge 1 + (n + 1)\alpha$.
