## Geometric Sequences

The __definition__ of a geometric sequence is, given $c \in \R : (x_n) = c^n \forall n \in \N$. If $|c|<1 \iff (x_n) \longrightarrow 0$, else $|c|\ge 1 \iff (x_n) \not\longrightarrow 0$.

__Lemma of Geometric Sequence Definition__, also known as Bernoulli's inequality, $(1 + \alpha)^n \ge 1 + \alpha n \forall \alpha \ge -1, \forall n \in \N$.

### Products and Quotients 

Suppose we have $x_n$ and $y_n : x_n \longrightarrow 0$ and $y_n \longrightarrow 0$. We may query, does $x_ny_n \longrightarrow 0$? Additionally, does $\frac{x_n}{y_n} \longrightarrow 0$?

It is true that $x_ny_n \longrightarrow 0$. It is __not__ true that its always the case that $\frac{x_n}{y_n}$. 

#### Theorem 3.15

Suppose $x_n \longrightarrow 0, y_n \longrightarrow 0$, then $x_ny_n \longrightarrow 0$. 

_Proof of Theorem 3.15_: Given $\epsilon > 0$, we want $|x_ny_n| < \epsilon$. This is distributive, i.e., $|x_n||y_n| < \epsilon$ (not to be confused with the triangle inequality). Given $|x_n||y_n| < \epsilon, \exists \epsilon' = \sqrt{\epsilon}$. 

Define $\epsilon' = \sqrt{\epsilon}, \epsilon >0$. Since $x_n \longrightarrow 0 \exists N_x \in \N : \forall n \in \N_x : |x_n| < \epsilon'$. Since $y_n \longrightarrow 0 \exists N_y \in \N : \forall n \in \N_y : |y_n| < \epsilon'$. 

Define $N = \max\{N_x, N_y\}$. Then, for $n > N: |x_ny_n| = |x_n||y_n| < \epsilon'\epsilon' = \sqrt{\epsilon}\sqrt{\epsilon} = \epsilon$.

#### Definition 3.18 - Bounded Sequences

A sequence $(c_n)_{n = 1}^{\infty}$ is bounded above if $\exists M \in \R : x_n \leq M \forall n \in \N$. A sequence $(x_n)_{n = 1}^{\infty}$ is bounded below if $\exists m \in \R : x_n \geq m \forall n \in \N$. Therefore a sequence $(x_n)_{n = 1}^{\infty}$, is bounded if and only if it is both bounded above and bounded below.

1. The sequence $x_n = \frac{1}{n}$ is bounded, i.e., it is bounded above and below. 
1. The sequence $x_n = n$ is __not__ bounded above, but is bounded below, i.e., $\not\exists M : M \ge x_n \forall n \in N$.

#### Theorem 3.21

If $x_n \longrightarrow 0$, and $y_n$ is bounded, then $x_ny_n \longrightarrow 0$.

This theorem generalises Theorem 3.15, as it applies to sequences that are both bounded and convergent.

&y_n \longrightarrow 0 \implies y_n \text{ is bounded}.\\
&\\

&M = \max\{|y_1|, |y_2|, \dots, |y_n|, |1|\} \implies |y_n| \le M \forall n \in \N.\\
\end{align*}$$
