## Introduction

An __ODE__ is a ordinary differential equation of an n-th order, and is a relation between a unknown function $y = y(x)$ of a __single independent variable__, say $x\in\mathbb{R}$, and the derivatives,

$$y' \equiv \frac{dy}{dx}, \dots, y^{(n)} \equiv \frac{d^n y}{dx^n}.$$

Symbolically, we can write any ODE in the form,

$$F(x, y, y', y'', \dots, y^{(n)}) = 0.$$

In essence, a ODE is an equation involving a function of a single real variable, and its derivatives. Note that the order of a ODE is the highest number of the differential.

### Definition

Any function $y = f(x)$ is defined in some interval $x \in (A, B)$, which when substituted into $F(x, y, y', \dots, y^{(n)}) = 0$ reduces the equation to a __identity__, and is called a __solution__ of the equation. Additionally, $(A, B)$ is known as the interval.

Differential equations have a direct relation to physics, computer science, chemistry etcetera. That is why these are extremely important to study.

In a practical sense, our singular variable $x$ is usually $t$ which is time. in this case, we use sightly different notation, 

$$\dot{y} = \frac{dy}{dt}, \ddot{y} = \frac{d^2 y}{dt^2}, \dots$$

### Example through Newton

Newton's second law is $f = ma$, where we can say that the point mass, $m$, is moving along the $y$ axis. By definition, $\dot{y} = v$ and $\ddot{y} = a$, which is $\dot{v} = a$. This means that we can express Newton's second law as a __second order differential equation__, 

$$m\ddot{y} = f(t, y, \dot{y}).$$

Where force, that's $f$, maybe time and velocity dependent as a function.

## Properties of First Order ODEs

### Explicit Solutions to First First Order ODEs

A first order ODE of __normal form__ is in general $y' = f(x, y)$. This allows us to develop a completely analytical solution. 

The __simplest case__ is where we have, $$y' = f(x).$$

The solution to such an example is simply $y = \int f(x) ~ dx + c, c \in \mathbb{R}$. 

#### Separable First Order ODEs

A separable ODE is one which is where our derivative is equal the product of a function of the independent variable and its following functions. I.e., for a first order separable ODE, the general form is, $$y' = f(x)g(y).$$

It is assumed that $f(x)$ and $g(y)$ are continuos on their respective domains.

If we set $g(y) = 0$ and get a set of solutions, say $y_1, y_2, y_3, \dots, y_k$, then we may state that $y(x) \in \{y_1, y_2, y_3, \dots, y_k\}$. This is simply saying that $g(y) = 0 \implies g(y(x)) = 0 $ for $y(x) \in \{y_1, y_2, y_3, \dots, y_k\}$. 

One may also take a different approach for a general solution, though it is not mathematically well defined and so we will soon ignore this method. 

Treating $\frac{dy}{dx}$ as a ratio of the difference of $x$ and $y$, we may,

$$\frac{dy}{dx} = f(x)g(y) \longrightarrow \frac{dy}{g(y)} = f(x)~ dx.$$

Now, we can integrate both sides to achieve the relation,

$$\int\frac{1}{g(y)} ~ dy =\int f(x)~ dx + c, \text{ for } c \in \mathbb{R}.$$

If we then create $H: \mathbb{R} \longrightarrow \mathbb{R} : H(y) \equiv \int \frac{1}{g(y)} ~ dy$, we can state $H(y) = \int f(x) ~dx + c$. Because we have a function $H$, $H^{-1}: \mathbb{R} \longrightarrow \mathbb{R}$, and its in this case there exists at least one inverse. __This is not a rigours justification__, though we will proceed with the following implication,

$$u = \int f(x) ~ dx + c\\ \\ H(y) = u \iff y = H^{-1}(u).$$

In light of this, we may develop a general solution which is,

$$y = H^{-1}\left(\int f(x) ~ dx + c\right).$$

One should note that there may exist multiple inverse functions of $H$, take the simple example of $H(y) = y^2, y^2 = u \implies y = \pm \sqrt{u}$ for $u > 0$. 

_Example One:_ Find a non constant solution to $y' = xy^2$

We have the equation $y' = xy^2 \equiv \frac{dy}{dx} = x y^2$, which can be put in form $\frac{dy}{y^2} = x~dx$. We need to integrate the expression,

$$\int \frac{1}{y^2} ~ dy = \int x ~ dx.$$

Finding $H(y) = \int \frac{1}{y^2} ~ dy$, we get $H(y) = (-2 + 1)^{-1}(y)^{-2 + 1} + c = -\frac{1}{y} + c$.

This gives us, $$-\frac{1}{y} = (1 + 1)^{-1}(x)^{1 + 1} + c = -\frac{1}{y} = \frac{x^2}{2} + c.$$

Given that $H(y) = -\frac{1}{y}$, set $u = -\frac{1}{y} : y = -\frac{1}{u} = H^{-1}(y).$

Given that the general solution is $y = H^{-1}\left(\int f(x) ~ dx + c\right)$, we see that,

$$y = -\frac{1}{\frac{x^2}{2} + c} = -\frac{2}{x^2} + c.$$

If there is such case when one cannot find a constant or general solution, we should simply state that the equation has a solution in an implicit form, given by the function of $y$.
