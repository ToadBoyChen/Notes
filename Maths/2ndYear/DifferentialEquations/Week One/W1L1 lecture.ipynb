_Example Two:_ $y'(t) = \frac{x}{y}$. Lets solve this.

$$\begin{align*}
&\frac{dy}{dx} = \frac{x}{y}.\\
&\implies y~dy = x~dx \implies \int y~ dy - \int x ~ dx + c\\
&\implies \frac{y^2}{2} = \frac{x^2}{2} + c \implies y^2 - x^2 = 2e.
\end{align*}$$

Taking a deeper look at the solution $y^2 - x^2 = 2e$, we can solve it as $y = \pm\sqrt{x^2 + 2c}$. These are __general explicit solutions__.

The equation $y^2 - x^2 = 2e$ is the __general implicit solution__, yes its singular in this case.

Equations of the form $y' = f(ax + by + c)$. This is not separable. $a,b,c\in \mathbb{R}$ and $f:\mathbb{R} \rightarrow \mathbb{R}$. It however is separable if $a = 0 \implies y' = f(by + c)$ or $b = 0 \implies y' = f(ax + c)$. What might be a suitable substitution? Introduce a new function $Z(x) = ax + by(x) + c \iff y(x) = \frac{Z(x) - ax - c}{b}$. This is considering the fact that $a, b \neq 0$. 

Plug in equations such that $\frac{Z'(x) - a}{b}$. Another substitution yields $\frac{Z'(x) - a}{b} = f(Z) \implies Z'(x) = a + bf(Z)$. This is now a separable ODE.

The solution is as follows,

$$\begin{align*}
&\frac{dZ}{a + bf(Z)} = dx \implies \int \frac{dZ}{a + bf(Z)} + c,\\

\end{align*}$$
