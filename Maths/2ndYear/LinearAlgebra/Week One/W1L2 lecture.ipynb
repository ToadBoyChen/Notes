## Systems of Linear Equations

These equations are concerned with solving for unknowns. 

### Definition

A linear equations in $n$ unknowns, $ x_1, \dots, x_n$ over $\mathbb{R}$ is an expression of the form, $$\alpha_1 x_1, \alpha_2 x_2, \dots, \alpha_n x_n = \beta, \alpha_1, \alpha_2, \dots, \alpha_n \in \mathbb{R}.$$

For $m, n \in \mathbb{N}$, a system of eqautions of $m$ linear equations in $n$ unknowns is the collection of equations of the form,

$$\begin{align*} 
&\alpha_{11} x_1 + \dots + \alpha_{mn} x_n = \beta_{m}\\
&\vdots\\
&\alpha_{m 1} x_1 + \dots + \alpha_{mn} x_{n} = \beta_{m}\\
&\\
&\text{where } \alpha_{ij}, \beta_j \in \mathbb{R}.
\end{align*}$$

A solution of the above system is a vector $c = \begin{pmatrix} \gamma_1 \\ \gamma_2 \\ \vdots \\ \gamma_n \end{pmatrix} \in \mathbb{R}^n$. 

Setting $x_j = \gamma_j$ for $j = 1, \dots, n$ all equations become satisfied, i.e., 

$$\forall~ i, \sum_{j = 1}^{n} \alpha_{ij} \gamma_{j} = \beta_{i}, \text{ for }\mathbb{R}.$$

the solution set of the system is the set of all solutions, where there maybe more than one solution vector.

__Matrix form__ of the above system is, 

$$Ax = b$$

where $A$ is a matrix, and the entries of an $n\times n$ matrix, $A$ may take the form,

$$\begin{pmatrix}
a_{11} & \dots & a_{1n} \\
\vdots & \ddots & \ddots \\
a_{m1} & \dots & a_{mn}
\end{pmatrix}$$

where $x = \begin{pmatrix} x_1 \\ \ddots \\ x_n\end{pmatrix}$ and $b = \begin{pmatrix} b_1 \\ \ddots \\ b_n\end{pmatrix}$. 

The solution set reads as, $\{c \in \mathbb{R}^n : Ac = b\}$.

The augmented matrix of a system in $\mathbb{R}^n$ takes form,

$$(A| B) = \begin{pmatrix} a_{11} & \dots & a_{1n} & \beta_1 \\ \vdots & \ddots & \vdots & \vdots \\ a_{m1} & \dots & a_{mn} & \beta_m \end{pmatrix}$$

### Definition

A linear system is __inconsistent__ if it has no real solutions.

Likewise, a system is __consistent__ if it has at __least one real solution__.

> This maybe a unique solution, or infinitely many solutions. 

It is easiest to solve solutions via row echelon form.

### Definition

__REF__ is defined be if, 

1. All zero rows are at the bottom of the matrix.
1. In each row, the first non zero entry from the left __must__ be $1$.
    1. This is known as a __leading one of the row__.
1. Each leading one is to the right og the leading $1$.

An even better way to solve a matrix is __reduced row echelon form__, or __RREF__. This is satisfied if additionally, if each leading one is the only non zero entry in a row.

A __leading variable__ is one which is obviously leading within a matrix, and __free ones__ are not. We can develop small and simple expressions for free variables. 

If we get some solution say,

$$X = \begin{pmatrix} x_1 \\ x_2 \\ x_3 \\ x_4\end{pmatrix} = \begin{pmatrix} -1 \\ 0 \\ 2 \\ 0\end{pmatrix} + \alpha\begin{pmatrix} -2 \\ 1 \\ 0 \\ 0\end{pmatrix} + \beta\begin{pmatrix} 7 \\ 0 \\ -2 \\ 1\end{pmatrix}.$$

this can be displayed in a solution set in __vector form__,

$$\left\{\begin{pmatrix} -1 \\ 0 \\ 2 \\ 0\end{pmatrix} + \alpha\begin{pmatrix} -2 \\ 1 \\ 0 \\ 0\end{pmatrix} + \beta\begin{pmatrix} 7 \\ 0 \\ -2 \\ 1\end{pmatrix} : \alpha, \beta \in\mathbb{R}\right\} \subseteq \mathbb{R}^4.$$

In essence, once we are in REF, its easy to find the solution set.
