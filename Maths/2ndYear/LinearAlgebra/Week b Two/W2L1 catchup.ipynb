## Systems of Equations

Two groups of systems are said to be __equivalent__ if they have the same solution set, which often takes form $\{a, b, c, d, \dots\}$.

Note that the basis for row operations is from the idea that,

1. We can interchange two equations,
1. We can multiply equations by a non zero scalar,
1. We can add multiples of other equations to a different equation.

__Row echelon form__ is where we have a upper triangular matrix, and __reduced row echelon form__ is where the leading non zero term is $1$ or $0$.

If we have an $m \times n$ matrix, it is said to be __overdetermined__ if it has more equations than unknowns, i.e., $m > n$, but __underdetermined__ if it has less equations than unknowns, i.e., $m < n$.

A system is said to be __homogenous__ if its equal to zero, and __inhomogeneous__ otherwise.

## Matrix Algebra

We write $A = (a_{ij})_{m \times n}$ to denote a $m \times n$ matrix whose $ij$-th entry is $a_{ij}$.

The zero matrix is a matrix of only zeros, and is denoted by $O_{m\times n}$.

For two matrices, $A = (a_{ij})_{m \times n}$ and $B = (b_{ij})_{n \times p}$, the produce $AB$ is $C = (c_{ij})_{m \times p}$ by $$c_{ij} = \sum_{k = 1}^n a_{ik}b_{kj}.$$

The identity matrix is a matrix who only contains zeros on the top left to bottom right diagonal. Denoted by $I_{m \times n}$.

Two matrices $A$ and $B$ and said to __commute__ if $AB = BA$.

If $AB = I$ and $BA = I$, then $B$ is the inverse of $A$.

The collection of all $n$-vectors is denoted by $\R^n$, and takes form,

$$\R^n = \left\{\begin{pmatrix}a_1\\a_2\\a_3\\\vdots\\a_n\end{pmatrix} \middle| a_1, a_2, a_3, \dots, a_n \in \R\right\}.$$

## Determinants
 
$$A = (a_{ij})_{n \times n} : \det(A) = \sum_{ i = 1}^{n} (-i)^{i + j}a_{ij}\det(A_{ij}).$$

note, we choose $j$ to be the easiest column to work with. 

$$A^{-1} = \frac{1}{\det(A)}\begin{pmatrix}
C_{11} & C_{21} & \dots & C_{n1}\\ 
C_{12} & C_{22} & \dots & C_{n2} \\ 
\vdots & \vdots & \ddots & \vdots \\
 C_{1n} & C_{2n} & \dots & C_nn\end{pmatrix}.$$

Where $C_{ji} = (-1)^{i + j}\det(A_{ji})$.

The __adjugate__ is the matrix of cofactor terms, so we may rewrite this formula as, $$A^{-1} = \frac{1}{\det(A)}\text{adj}(A).$$

## Linear Algebra

### Vector Spaces

#### Definition of a Vector Space

A __Vector Space__ is a non empty set $V$ where there are two operations called _addition_ and _scalar multiplication_, such that the following axioms hold for $u, v, w \in V$, and $\alpha, \beta \in \R$,

1. $u + v \in V$,
1. $\alpha u \in V$,
1. $u + v = v + u$,
1. $u + (v + w) = (u + v) + w$,
1. $u + 0 = 0 + u = u$,
1. $u + (-u) = u + (-1)(u) = 0$,
1. $\alpha(u + v) = \alpha u + \alpha v$,
1. $(\alpha + \beta)u = \alpha u + \beta v$,
1. $(\alpha \beta)u = \alpha(\beta u)$,
1. $1u = 1(u) = u$.

$V$ refers to the universal set fo the vector space, where its elements are vectors. We work with __real vector spaces__ and __complex vectors spaces__.

### Sub Spaces
A subspace of a vector space is in some sense a subset of a $V$, i.e., $H \subseteq V$, making $H$ almost a subspace. A subspace itself is obviously a subspace, and this can allow us to prove vector spaces exist by proving vector subspaces, which is __much easier__. The only axioms a subspace must obey are the __closure axioms__, this being two axioms easily proven. 

#### Definition of a Vector Subspace
Let $H \neq \empty : H \subseteq V$. $H$ is a subspace $\iff u, v \in H : u + v \in H$ and $u \in H, \alpha in \R : \alpha u \in H$. 

It follows by logic that $H$ if inheriting addition and scalar multiplication from $V$ that $H$ it itself a vector subspace.

Some important cases of subspaces is that given $V$, $\{V\}$ and $\{0\}$ are themselves trivial __but valid subspaces of__ $V$. The Zero subspace is obviously $\{0\}$, and all __proper subspaces__ are subspaces which are not $\{0\}$ nor $\{V\}$.

_Non Example_: $H = \{(r^2, s, r)^T : r, s \in \R\}$. This must satisfy the closure laws to be considered a subspace and therefore in its own respect a vector space but,

$$\begin{pmatrix}1 \\ 0 \\ 1\end{pmatrix} \in H : 2\begin{pmatrix}1 \\ 0 \\ 1\end{pmatrix} \notin H.$$

Let $A \in \R^{m \times n} : N(A) = \{X \in \R^n : Ax = 0\}$. This is known as the __null space__ of $A$. It should be noted that $A \in \R^{m \times n}$ then $N(A)$ is a subspace of $\R^n$.

The null space can be though of as the set of all solutions to a homogeneous system. An example may look as such:

$$A = \begin{pmatrix} 
-3 & 6 & -1 & 1 & -7 \\
1 & -2 & 2 & 3 & -1 \\
2 & -4 & 5 & 8 & -4\end{pmatrix},\\ \\ \implies N(A) = \left\{\alpha\begin{pmatrix}2\\1\\0\\0\\0\end{pmatrix} + \beta\begin{pmatrix}1\\0\\-2\\1\\0\end{pmatrix} + \gamma\begin{pmatrix}-3\\0\\2\\0\\1\end{pmatrix} \middle| \alpha, \beta, \gamma \in \R\right\}.$$

### The Span of a Set of Vectors
We will look into how we can construct subspaces from a collection of vectors. 

#### Definition of the Span of a Set of Vectors
Let $v_1, \dots, v_n$ be vectors in a vector space $V$. The set of all linear combinations of $v_1, \dots, v_n$ is called the __span__ of $\{v_1, \dots, v_n\}$. We denote this span as $\text{span}\{v_1, \dots, v_n\}$. More formally, 

$$\text{Span}(v_1, \dots, v_n) = \{\alpha_1 v_1 + \dots + \alpha_n v_n : \alpha_1, \dots, \alpha_n \in \R\}.$$
