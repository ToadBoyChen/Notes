__Fact__: The $\gcd(n, 1)$ for $n \in \mathbb{Z}$ is $\gcd(n, 1) = 1$.

_Take the example of_: $\gcd(99, 72) = ?$ The calculation is as following,

$$\begin{align*}
\gcd(99, 72)& \implies 99 = 1(72) + 27\\
& 72 = 2(27) + 18 \implies \gcd(99, 72) = \gcd(72, 27)\\
& 27 = 1(18) + 9 \implies \gcd(72, 27) = \gcd(27, 18)\\
& 18 = 2(9) + 0 \implies \gcd(99, 72) = 9.
\end{align*}$$

Once again, $a, b \in \mathbb{Z}$ with $b \in \mathbb{N} : a = qb + r$, with $0 \leq r < b$, then $\gcd(a, b) = \gcd(b, r)$. We can recall that $b$ must be bigger than $0$ by the statement $0 /leq r < b$. This __is__ Euclid's algorithm.

_Proof_: $\gcd(b, r) = \gcd(a, b)$. To prove this, we must prove,

1. $\gcd(b, r) \mid a$ and $\gcd(b, r) \mid b$, 
1. If $e \mid a$ and $e \mid b \implies e \leq d$.

Following this, $\gcd(b, r) = d$ implies that $d \mid r$ and $d \mid b$. And using $a = qb + r \implies d\mid a$, as $r$ is a multiple of $d$ and $b$ is a multiple of $d$. I.e., $d \mid a$ as $qb + r \mid d$.

Now to prove 2., Recall that $\gcd(b, r) = d$. Is follows that $e \mid b $ and $e \mid a$ (by hypothesis) $\implies e \mid -qb + a \iff e \mid r$. It shows that $e \mid b$ and $e \mid r$, so that $e \leq d$. _QED_

## Bezout's Identity

### Definition

Let $a, b \in \mathbb{Z}$, then $\exists ~ r, s \in \mathbb{Z} : ar + bs = \gcd(a, b)$.

We are given two numbers by which the gcd is able to be expressed in terms of the $a, b$ as some multiples of $r, s$.

### Examples

_Example One:_ $a = 6, b = 5$, note that $1(6) -1(5) = 1$ and that $\gcd(6, 5) = 1$. How should one find these two multiples which satisfies Bezout's identity? 

_Example Two:_ $\gcd(12, 20) = 4$. How may we find $r, s$? $$r(12) + s(20) = 4 \implies 2(12) + -1(20) = 4.$$

_Example Three:_ $\gcd(-11, 15) = 1$. $$r(-11) + s(15) = 1 \implies 4(-11) + 3(15) = 1.$$

To actually find $r, s$ we must calculate the gcd and work in reverse. 

_Example One - Continued:_ $\gcd(6, 5) = 1 \implies 6 = 1(5) + 1 \implies 1(6) + -1(5) = 1$.

_Example Two - Continued:_ $\gcd(12, 20) = 4$. 

$$\begin{align*}
20&=1(12) + 8\\
12&= 1(8) + 4\\
8 &= 2(4) + 0
\end{align*}$$

Now working backwards, we get,

$$\begin{align*}
4 &= 12 - 8\\
8 &= 20 - 12\\
4 &= 12 - 20 + 12\\
&= 2(12) - 20.
\end{align*}$$

It's therefore given that $r = 2$ and $s = -1$.

_Example Three - Continued:_ $\gcd(-11, 15) = 1$. 

$$\begin{align*}
-11 &= -1(15) + 4\\
15 &= 3(4) + 3\\
4 &= 1(3) + 1\\
3 &= 3(1) + 0 \implies \gcd(-11, 15) = 1.
\end{align*}$$

Now working backwards, we get,

$$\begin{align*}
1 &= 4 - 3\\
3 &= 15 - 3(4)\\
4 &= -11 + 15\\
\implies 1 &= -11 + 15 - 15 + 3(4)\\
1 &= -11 + 3(-11 + 15)\\
1 &= 4(-11) + 3(15).  
\end{align*}$$

It's therefore given that $r = 4$ and $s = 3$.

### Following Lemma

Let $a, b \in \mathbb{Z}, d \in \mathbb{N}$. Then the following two statements are equivalent. 

1. The equation $ax + by = d$ has solutions in $x,y \in \mathbb{Z}$.
1. $\gcd(a,b) \mid d$.

$$ax + by = d \text{ for }x,y\in\mathbb{Z} \iff \gcd(a, b) \mid d.$$

_Rationale One:_ can we find integer solutions in the equation $2x + 4y = 3$? Note that the number of the solutions is not infinite as we are bound by integers. Given that both $2$ nor $4$ divides $3$, we have __no__ solutions. I.e., $\gcd(2,4) = 2 \nmid 3 \not{\implies} 2x + 4y = 3$.

_Rationale Two:_ can we find integer solutions in the equation $5x + 15y = 25$? __Yes__, because of our lemma $ax + by = d \text{ for }x,y\in\mathbb{Z} \iff \gcd(a, b) \mid d$.

#### Proof of Lemma

Prove that,

1. $ax + by = d \implies \gcd(a, b) \mid d$,
1. $\gcd(a, b) \mid d \implies ax+by = d$.

Let $g = \gcd(a, b) : g \mid a $ and $ g \mid b \implies g \mid ax + by = d$.

Given Bezout's identity, $\exists~ r,s\in\mathbb{Z} : ar + bs = \gcd(a, b) = g$. As $g \mid d \implies ~\exists~ q\in\mathbb{Z} : d = qg$. If $x = rq$ and $y = sq$ then $ax + by = arq + bsq = q(ar+bs) \implies gq = d$.

## Primes and factorisation

### Definition of Prime

#### Def One

Given $b \in \mathbb{N}$ is prime if b can be divided by __only__ $\pm 1$ and $\pm b$.

#### Def Two

Let $a,b\in\mathbb{Z}: b$ is prime. The following are equivalent,

1. $b \mid ab$,
1. $b \mid a ~\lor~ b \mid b$.

#### Fundamental Theorem of Arithmetic

Every $ n \in \mathbb{N}$ can be written as a produce of prime numbers. Moreover, the way of writing the product is unique, up to reordering. 

_Example One:_ $n = 15 = 3 \times 5 = 5 \times 3$.

_Example Two:_ $n = 12 = 2 \times 2 \times 3 - \dots$.

Reordering refers to the fact that we do __not__ care about order, but the actual numbers are unique.
