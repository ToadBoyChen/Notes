## Motivation

### Fermat's Last Theorem

Find solutions in $(x,y)\in\mathbb{Q}$ such that $x^3 + y^3 = 1$. 

Some solutions may include $1^3 + 0^3 = 1$. Its actually a known fact that the solutions $(1,0)$ and $(0,1)$ are the only ones to exist.

One should look into _Langland's program_, which is in relation to Fermat's last theorem. This also relates another topic known as _Arithmetic geometry_. Fermat's last theorem (proven) shows that,

$$\text{Only one trivial solution: }\begin{cases}
x^3 + y^3 &= 1\\
x^4 + y^4 &= 1\\
x^5 + y^5 &= 1\\
&\vdots
\end{cases}$$

### Goldbach's Conjecture

Goldbach observed that every even number is a sum of two prime numbers.

_Example_: $$4 = 2 + 2\\ 6 = 3 + 3\\ 8 = 5 + 3\\ \vdots$$.

__This is still an open problem__.

### Diophantine Approximations

As we know, there are irrational numbers, where we cannot write a number as a ratio, i.e., $\sqrt{2} = \frac{a}{b}$ has no such $a$ or $b$ to fully equal $\sqrt{2}$. Though we may __approximate__ $\sqrt{2}$, which is known as a Diophantine approximation. An approximation for root $2$ is $$\sqrt{2} \approx \frac{141}{100} \text{ or } \frac{1414}{1000} \dots$$

## Chapter One

### Notation

#### Number Sets

The integer set is $\mathbb{Z} = \{\dots,-2, -1, 0, 1, 2, \dots\}$. Z is used because of Zahlem, the german work for number. 

The natural set is $\mathbb{N} = \{1, 2, 3, 4, \dots\}$.

The natural set __with zero__ is $\mathbb{N}_0 = \{0,1,2,3,\dots\}$.

#### Operations

For $a,b\in\mathbb{Z}$, we say that $a \mid b \iff qa = b, q\in\mathbb{Z}$.

> We may state things like $2 \mid 6$ as $3(2) = 6$.

### Euclid's Algorithm

Given $a\in\mathbb{Z}$ and $b\in\mathbb{N}$ $\exists~ r,q\in \mathbb{Z}: 0 \leq r < b$ and $a = qb + r$. $r, q$ are __unique__.

> Example: Take two numbers $(7, 3)$ such that we get $7 = 2(3) + 1$. This implies that the $\gcd(7, 3) = 1$.

> Example: Take two numbers $(15, 9)$ such that we get $15 = 1(9) + 6$. This is further iterated by $9 = 1(6) + 3$ and then $6 = 2(3) + 0$. This implies that the $\gcd(15, 9) = 3$.

### Greatest Common Divisor

For $a, b \in \mathbb{Z}$, we define $d = \gcd(a, b)$ if $d$ is a common divisor and $d$ is the largest among all the common divisors. 

We say that $a, b$ are coprime to each other if $\gcd(a,b) = 1$.

#### Lemma

Let $a = qb + r, 0 \leq r < b$. The $\gcd(a,b) = \gcd(b, a)$.
