## Congruences and modular Arithmetic

Let $n \in \N$. We say that $a, b \in \Z$ are congruent modulo $n$ if $n\mid(a - b)$ or $a-b = k(n), k \in \N$, and we write $a \equiv b \mod n$, or $a \equiv_n b$ or $a \equiv b$.

The congruence operation is in fact an equivalence relation; there are $n$ equivalence classes corresponding to $0, 1, \dots, n - 1$, and are called congruence classes modulo $n$. We denote the classes by $[a]_n$, that is to say that $[a]_n = \{z \in \Z : z \equiv a \mod n\} = \{\dots, a - 2n, a - n, a, a + n, a + 2n, \dots\}$.

By definition, any member of the congruence class $[a_n]$ can also represent the class, i.e., $a \equiv b \mod n$.

Some helpful properties are that $[a]_n + [b]_n = [a + b]_n$, and that $[a]_n \times [b]_n = [a \times b]_n$.

One should note that representatives shouldn't matter when dealing with these basic operations. $a \equiv a' \mod n' \implies [a] + [b] = [a'] + [b']$? i.e., does it matter if we have say the equation $[2]_5 + [3]_5 = [7]_5 +  [8]_5$? No it doesn't.

_Example_, here is the set $\Z/4\Z$:

$$\begin{align*}
[0]_4 &= \{\dots, -12, -8, -4, 0, 4, 8, 12, \dots\}\\
[1]_4 &= \{\dots,-11,-7,-3,1,5,9,13,\dots\}\\
[2]_4 &= \{\dots,-10,-6,-2,2,6,10,14,\dots\}\\
[3]_4 &= \{\dots,-9,-5,-1,3,7,11,15,\dots\}\\
\end{align*}$$

We can also construct an addition table,

$$\begin{array}{c|cccc}
+ & [0]_4 & [1]_4 & [2]_4 & [3]_4 \\
\hline
[0]_4 & [0]_4 & [1]_4 & [2]_4 & [3]_4 \\
[1]_4 & [1]_4 & [2]_4 & [3]_4 & [0]_4 \\
[2]_4 & [2]_4 & [3]_4 & [0]_4 & [1]_4 \\
[3]_4 & [3]_4 & [0]_4 & [1]_4 & [2]_4 \\
\end{array}$$

And a multiplication table,

$$\begin{array}{c|cccc}
\times & [0]_4 & [1]_4 & [2]_4 & [3]_4 \\
\hline
[0]_4 & [0]_4 & [0]_4 & [0]_4 & [0]_4 \\
[1]_4 & [0]_4 & [1]_4 & [2]_4 & [3]_4 \\
[2]_4 & [0]_4 & [2]_4 & [4]_4 & [2]_4 \\
[3]_4 & [0]_4 & [3]_4 & [2]_4 & [0]_4 \\
\end{array}$$

#### Proposition

If $p$ is a prime, then $\mathbb{F}_p$ is a field.

We should remark that many different congruence classes are not fields, but rings instead. This is simply meaning we can perform our addition and multiplication but not division. Our proposition simply asserts that the integers modulo $p$ form a field, or $[a]_p$ has a multiplicative inverse, for all $a \neq 0$.

#### Fermat's Little Theorem
Let $p$ be a prime number. Then $z^p \equiv z \mod p$ for any $z \in \N$.

### Congruence Equations

#### Proposition

Let $a, n$ be natural and $b \in \Z$. $ax \equiv d \mod n \iff \gcd(a, n) | d$.

#### Chinese Remainder Theorem
This theorem allows us to solve simultaneous congruences to different moduli. We say that $n$ and $m$ are coprime if $\gcd(n, m) = 1$.

Let $m, n$ be coprime. Then there is a solution to the simultaneous congruence equations,

$$\begin{align*}
x &\equiv a \mod m,\\
x &\equiv b \mod n.
\end{align*}$$

More generally,

$$\begin{align*}
x &\equiv a_i \mod n_i,\\
\end{align*}$$

where $n_1, \dots, n_i$ are pairwise coprime.

#### Theorem 
There are infinitely many primes congruent to $-1 \mod 4$. 

## Euler's Totient Function and Primitive Roots

### Euler's Totient Function

#### Definition

Euler's totient function, or $\phi$-function is the function $\phi : \N \rightarrow \N$ by the amount of natural numbers such that $\gcd(z, n) = 1$ for $1 \le z \le n$. 

_Example, if $p$ is prime, $\phi(p) = p - 1$ because $1, 2, 3, \dots, p-1$ are all coprime to $p$._

_Example, $\phi(8) = 4$, the odd numbers are coprime to $8$ but the even ones are not._

If $R$ is a commutative ring with identity, then the element $r \in R$ is said to be a unit if there exists $s$ such that $rs = 1 \in R$. The units form a group under multiplication.
