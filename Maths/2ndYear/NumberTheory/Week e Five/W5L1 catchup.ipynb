## Euler's Totient Function and Primitive Roots

### Euler's Totient Function

__Definition__ Euler's Totient Function, or the $\phi$ function is $\phi : \N \longrightarrow \N$ by $n \in \N \mapsto |1 \le z \le n| : \gcd(z, n) = 1$.

________________________________________________________________
__Example__ If $p$ is prime then $\phi(p) = p - 1$, as $1, 2, 3, \dots, p-1$ are all co-prime to $p$.

__Example__ If $n = 8$, we have $\phi(8) = 4$ as $1, 3, 5, 7$ are all co-prime to $8$.

__Definition__ If $R$ is a commutative ring with an identity, then $r \in R$ is said to be a unit if $\exists s \in R : rs = 1$. The units in $R$ form a group under multiplication, denoted by $R^\times$.

#### Proposition 14

The number $|(\Z / n\Z^\times)|$ of elements from the group $(\Z / n\Z^\times)$ of units in $\Z / n\Z$ is $\phi(n)$. 

_Proof_: Let $[z]$ be a unit in $\Z / n\Z \iff \gcd(z, n) = 1$. It then follows that $\exists r, s \in \Z : zr + ns = 1$, by Bezout's identity. It also follows that $zr \equiv 1 \mod N \iff [z][r] = [1]$, for $[z]$ being the unit. Conversely suppose that $[z]$ is a unit (as we are proving a implicit iff statement, which is stronger than a normal implication). $\exists [r] \in \Z : [z][r] = [zr] = [1]$. It therefore follows that $zr \equiv 1 \mod n \implies zr + ns = 1, r \in \Z$. If we let $d = \gcd(z, n)$, by definition $d$ divides $z$ and $n$, and therefore $zr + ns$. Therefore $d = 1$. 

#### Theorem 15

Let $n$ be a positive integer and $z$ be an integer such that $\gcd(z, n) = 1$. Then $z^{\phi(n)}\equiv 1 \mod n$. 

_Proof_: (1) Let $N = \phi(n)$ and let $z_1, \dots, z_N$ be the integers in the set $\{0, 1, \dots, n\}$ that are relatively prime to $n$. 

> (1) We define $N$ as $\phi(n)$ to simply shorten the proof. We then define some general integers $z_1, \dots, z_N$ as to be the same integers in $\{0, 1, \dots, n\}$ such that $\gcd(z_i, n) = 1$, for $i = 1, \dots, N$.

(2) Consider the set $\{r_1, \dots r_N\}$ where $r_j$ is defined to be the residue $1 \le r_j \le n - 1$ of the product $zz_j$ when divided by $n$. We may state that $\{r_1, \dots, r_N\} = \{z_1, \dots, z_N\}$. 

> (2) We are saying that the remainder of $\frac{zz_j}{n}$ corresponds to the same integer $r_j$ for all $j = 1, \dots, N$.

(3) Let us assume that $r_i = r_j, 1 \le i,j \le N \implies zz_i \equiv zz_j \mod n \implies i \equiv j \mod n$.

> (3) Here we state that if there are some terms, say $r_i$ and $r_j : r_i = r_j$, it means that $i = j$. This means we have a one to one mapping. We may $zz_i \equiv zz_j \mod n \iff z_i \equiv z_j \mod n$ as there exists $z^{-1}$ because $z$ is coprime to $n$. 

(4) Since $\{zz_1, zz_2, \dots, zz_N\} \equiv \{r_1, r_2, \dots, r_N\} = \{z_1, z_2, \dots, z_N\} \mod n$, we get $$z^N \prod_{j = 1}^{N}z_j = \prod_{j = 1}^{N}zz_j \equiv \prod_{j = 1}^{N}z_j.$$

> (4) We are saying that the product of all $z_j$ modulo $n$ is equal to the product of all $z_j$.

(5) Therefore, $z^N \equiv 1 \mod n$.

#### Corollary 16

$p$ is prime. Then $z^p \equiv z \mod p, z \in \Z$. 

#### Theorem 17

1. If $p$ is prime and $r > 0$, then $\phi(p^r)$ = p^{r - 1}(p-1).
1. If $\gcd(k, \ell) = 1 \iff \phi(k\ell) = \phi(k)\phi(\ell)$.
1. If $n = \prod_{j = 1}^{s}p_j^{r_j}$, where $p_1, \dots, p_s$ are distinct primes and $r_1, \dots, r_s > 0$, then,

$$\phi(n) = \prod_{j = 1}^{s}p_j^{r_j - 1}(p_j - 1) = n\prod_{j = 1}^{s}\left(1 - \frac{1}{p_j}\right).$$

#### Proposition 18

Let $d$ be the divisor of $n$. Then the number of integers $z$ with $1 \le z \le n$ and $\gcd(z, n) = d$ is $\phi\left(\frac{n}{d}\right)$. 

#### Definition

Let $n \in \N$. If there exists a positive integer $d$ such that $z^d \equiv 1 \mod n$, then the order of $z \mod n$ is the smallest of all integers $d$. the order of $z$ may be defined as the smallest integer $d$ such that $[z]_n^d = [1]_n$ in the set $\Z / n\Z$ of congruence classes modulo $n$.

#### Lemma 19

Suppose $z$ has order $d$. If $z^e \equiv 1 \mod n$, then $d | e$.

#### Proposition 20

For integers $z \exists d \in \N : z^d \equiv 1 \mod n \iff \gcd(z, n) = 1$. If so, the order of $z$ divides $\phi(n)$.

### Primitive Roots

#### Definition

let $p$ be a prime number. An integer $z$ is said to be a primitive root $\mod p$ if $z$ has order $p - 1 (\mod p)$. Note that since $\phi(p) = p - 1$, a primitive root has the maximum possible order.
